```python
## Prepare image path in txt file
```

```python
import os
```

```python
train_img_path = "/content/drive/MyDrive/Machine Learning/YOLO v7/images/train"
val_img_path = "/content/drive/MyDrive/Machine Learning/YOLO v7/images/val"
```

```python
#training images

with open('train.txt', 'a+') as f:
  img_list = os.listdir(train_img_path)
  for img in img_list:
    f.write(os.path.join(train_img_path,img+'\n'))
print("Done")
```

```python
## Validation Image
with open('val.txt', 'a+') as f:
  img_list = os.listdir(val_img_path)
  for img in img_list:
    f.write(os.path.join(val_img_path,img+'\n'))
print("Done")
```
